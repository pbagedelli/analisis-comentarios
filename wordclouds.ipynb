```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
comentarios = pd.read_csv('./data/commentarios_lanacion-ultimas100.csv', index_col  = 0)
```

```python
lanacion = pd.read_csv('./data/lanacion_notas.csv', index_col  = 0)
infobae = pd.read_csv('./data/infobae_notas.csv', index_col  = 0)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('spanish'))
stop_words.update(["si"])
```

```python
infobae.seccion.unique(), lanacion.seccion.unique()
```

```python
economiaL = lanacion.loc[lanacion.seccion == 'Economía']
economiaI = infobae.loc[infobae.seccion == 'economia']

politicaL = lanacion.loc[lanacion.seccion == 'Política']
politicaI = infobae.loc[infobae.seccion == 'politica']
sociedadL = lanacion.loc[lanacion.seccion == 'Sociedad']
sociedadI = infobae.loc[infobae.seccion == 'sociedad']
```

```python
# Start with one review:
text = ' '.join(politicaL.titulo)

# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,  stopwords=stop_words, background_color='white' ).generate(text)
 
# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
