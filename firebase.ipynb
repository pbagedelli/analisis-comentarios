```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore

cred = credentials.Certificate("comentarios-f3c78-firebase-adminsdk-nw1oe-1cd65fcf9e.json")
firebase_admin.initialize_app(cred)
```

```python
db = firestore.client()
```

```python
results = db.collection(u'comentarios').get()
```

```python
comentarios = []
for result in results:
    comentarios.append(result.to_dict())
```

```python
import pandas as pd
todo = pd.DataFrame(comentarios)
```
