```python
import re
misoginos = df.loc[df.texto.str.contains(glosariosRe.misoginos, flags = re.IGNORECASE, na = False)]
racistas = df.loc[df.texto.str.contains(glosariosRe.racistas, flags = re.IGNORECASE, na = False)]
xenofobos = df.loc[df.texto.str.contains(glosariosRe.xenofobos, flags = re.IGNORECASE, na = False)]
ideologicos = df.loc[df.texto.str.contains(glosariosRe.ideologicos, flags = re.IGNORECASE, na = False)]
violentos = df.loc[df.texto.str.contains(glosariosRe.violentos, flags = re.IGNORECASE, na = False)]
homofobicos = df.loc[df.texto.str.contains(glosariosRe.homofobicos, flags = re.IGNORECASE, na = False)]
periodistas = df.loc[df.texto.str.contains(glosariosRe.periodistas, flags = re.IGNORECASE, na = False)]
```

```python
misoginos.sort_values(by = 'toxicity', ascending= False)
```

```python
misoginos.sort_values(by = 'toxicity', ascending= False).to_csv('./data/misoginos_infobae.csv')
racistas.sort_values(by = 'toxicity', ascending= False).to_csv('./data/racistas_infobae.csv')
xenofobos.sort_values(by = 'toxicity', ascending= False).to_csv('./data/xenofobos_infobae.csv')
ideologicos.sort_values(by = 'toxicity', ascending= False).to_csv('./data/ideologicos_infobae.csv')
violentos.sort_values(by = 'toxicity', ascending= False).to_csv('./data/violentos_infobae.csv')
homofobicos.sort_values(by = 'toxicity', ascending= False).to_csv('./data/homofobicos_infobae.csv')
periodistas.sort_values(by = 'toxicity', ascending= False).to_csv('./data/periodistas_infobae.csv')
```
