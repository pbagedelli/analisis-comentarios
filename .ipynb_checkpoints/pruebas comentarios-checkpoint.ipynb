```python
import glosarios
import pandas as pd
```

```python
df = pd.read_csv('./data/commentarios_lanacion.csv', index_col = 0)
```

```python
#df = pd.read_csv('./data/commentarios_infobae.csv', index_col = 0)
```

```python
import re
misoginos = df.loc[df.texto.str.contains('|'.join(glosarios.misoginos), flags = re.IGNORECASE, na = False)]
racistas = df.loc[df.texto.str.contains('|'.join(glosarios.racistas), flags = re.IGNORECASE, na = False)]
xenofobos = df.loc[df.texto.str.contains('|'.join(glosarios.xenofobos), flags = re.IGNORECASE, na = False)]
ideologicos = df.loc[df.texto.str.contains('|'.join(glosarios.ideologicos), flags = re.IGNORECASE, na = False)]
violentos = df.loc[df.texto.str.contains('|'.join(glosarios.violentos), flags = re.IGNORECASE, na = False)]
homofobicos = df.loc[df.texto.str.contains('|'.join(glosarios.homofobicos), flags = re.IGNORECASE, na = False)]
```

```python
misoginos.to_csv('./data/misoginos.csv')
racistas.to_csv('./data/racistas.csv')
xenofobos.to_csv('./data/xenofobos.csv')
ideologicos.to_csv('./data/ideologicos.csv')
violentos.to_csv('./data/violentos.csv')
homofobicos.to_csv('./data/homofobicos.csv')
```
