```python
import requests
import json
```

```python
import pandas as pd
```

```python
df= pd.read_csv('./data/autores_infobae.csv', index_col = 0)
```

```python
def genero(nombre):
    return json.loads(requests.get('https://api.genderize.io/?name=' + nombre).text)['gender']
```

```python
df['nombre'] = df.loc[:,'0'].dropna().str.extract('(\w+)')
```

```python
df['genero'] = df.nombre.dropna().apply(genero)
```

```python
df.rename(columns= {'0': 'autor'}).to_csv(('./data/autores_infobae.csv'))
```
