```python
import pandas as pd
df = pd.read_csv('./data/lanacion_notas.csv', index_col = 0)
comentarios = pd.read_csv('./data/commentarios_lanacion.csv', index_col = 0)
```

```python
autores = pd.read_csv('./data/lanacion_toxicity_autor.csv', index_col = 0)
```

```python
autores = autores.merge(df.groupby('autor').count()[['titulo']], on = 'autor')
```
